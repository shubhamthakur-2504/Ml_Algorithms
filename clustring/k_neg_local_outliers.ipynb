## k neighbor local outliers

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
```

```python
x,y = make_circles(n_samples=5000,factor=0.30,noise=0.1)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
local_outlier = LocalOutlierFactor(n_neighbors=20, algorithm="auto", contamination=0.2)
y_pred = local_outlier.fit_predict(x)
# y_pred = local_outlier.predict(x)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x[:,0],x[:,1],c=y_pred)
```
