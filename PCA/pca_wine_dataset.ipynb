## PCA on Wine Dataset

```python
## importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
## loading the dataset
wine = load_wine()
```

```python
wine.keys()
```

```python
print(wine['DESCR'])
```

```python
# creating a dataframe
wine_data = pd.DataFrame(data=wine['data'],columns=wine['feature_names'])
wine_data.head()
```

```python
# standardizing the data
scaler = StandardScaler()
wine_scaled = scaler.fit_transform(wine_data)
```

```python
wine_scaled
```

```python
# applying pca
pca = PCA(n_components=2)
pca_wine = pca.fit_transform(wine_scaled)
pca_wine
```

```python
# visualizing the data
plt.figure(figsize=(8,6))
plt.scatter(pca_wine[:,0],pca_wine[:,1],c=wine['target'],cmap='plasma')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
