### without hypermeter tunning 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

%matplotlib inline

x,y = make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=15)
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.30,random_state=42)

logistic = LogisticRegression()
logistic.fit(x_train,y_train)

y_pred = logistic.predict(x_test)
score=accuracy_score(y_test,y_pred)
classification_report=classification_report(y_test,y_pred)
confusion_matrix=confusion_matrix(y_test,y_pred)
print(score)
print(classification_report)
print(confusion_matrix)
```

### with gridsearchcv

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']

prams=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()

grid=GridSearchCV(estimator=model,param_grid=prams,cv=cv,n_jobs=-1,scoring='accuracy')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(x_test)
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(accuracy_score(y_test,y_pred))
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_test,y_pred))
```

### randomizedsearchcv

```python
from sklearn.model_selection import RandomizedSearchCV

model=LogisticRegression()
random=RandomizedSearchCV(estimator=model,param_distributions=prams,cv=cv,n_jobs=-1,scoring='accuracy')


random.fit(x_train,y_train)
```

```python
random.best_params_
```

```python
random.best_score_
```

```python
y_pred=random.predict(x_test)
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(accuracy_score(y_test,y_pred))
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_test,y_pred))
```
