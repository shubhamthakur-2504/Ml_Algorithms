### Cleaning dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_set = pd.read_csv('Travel.csv')
```

```python
data_set.head()
```

```python
data_set['Gender'].value_counts()
```

```python
data_set['Gender'].replace('Fe Male', 'Female', inplace=True)
```

```python
data_set['Gender'].value_counts()
```

```python
data_set.head()
```

```python
data_set['MaritalStatus'].value_counts()
```

```python
data_set['MaritalStatus'].replace('Single','Unmarried',inplace=True)
```

```python
data_set["MaritalStatus"].value_counts()
```

```python
data_set.head()
```

```python
data_set['TypeofContact'].value_counts()
```

```python
data_set['Occupation'].value_counts()
```

```python
data_set['ProductPitched'].value_counts()
```

```python
data_set['Designation'].value_counts()
```

```python
data_set.isnull().sum()
```

```python
feature = data_set.columns
feature_with_nan = [feature for feature in data_set.columns if data_set[feature].isnull().sum() >= 1]
```

```python
for feature in feature_with_nan:
    print(feature, np.round(data_set[feature].isnull().mean()*100,5))
```

```python
data_set.dtypes
```

```python
data_set[feature_with_nan].select_dtypes(exclude='object').describe()
```

```python
data_set.fillna({'Age':data_set['Age'].median()}, inplace=True)
```

```python
data_set['Age'].isnull().sum()
```

```python
data_set.head()
```

```python
data_set.fillna({'TypeofContact':data_set['TypeofContact'].mode()[0]}, inplace=True)
```

```python
data_set['TypeofContact'].isnull().sum()
```

```python
data_set.fillna({'DurationOfPitch':data_set['DurationOfPitch'].median()}, inplace=True)
```

```python
data_set['DurationOfPitch'].isnull().sum()
```

```python
data_set.fillna({'NumberOfFollowups':data_set['NumberOfFollowups'].mode()[0]}, inplace=True)
```

```python
data_set['NumberOfFollowups'].isnull().sum()
```

```python
data_set.fillna({'PreferredPropertyStar':data_set['PreferredPropertyStar'].mode()[0]}, inplace=True)
```

```python
data_set['PreferredPropertyStar'].isnull().sum()
```

```python
data_set.fillna({'NumberOfTrips':data_set['NumberOfTrips'].median()}, inplace=True)
```

```python
data_set['NumberOfTrips'].isnull().sum()
```

```python
data_set.fillna({'NumberOfChildrenVisiting':data_set['NumberOfChildrenVisiting'].mode()[0]}, inplace=True)
```

```python
data_set['NumberOfChildrenVisiting'].isnull().sum()
```

```python
data_set.fillna({'MonthlyIncome':data_set['MonthlyIncome'].median()},inplace=True)
```

```python
data_set['MonthlyIncome'].isnull().sum()
```

```python
data_set.isnull().sum()
```

```python
data_set.drop(['CustomerID'], axis=1, inplace=True)
```

```python
data_set.head()
```

```python
data_set['TotalVisiting'] = data_set['NumberOfPersonVisiting'] + data_set['NumberOfChildrenVisiting']
data_set.drop(['NumberOfPersonVisiting','NumberOfChildrenVisiting'], axis=1, inplace=True)
data_set.head()
```

```python
data_set.to_csv('Holiday_cleaned_dataset.csv', index=False)
```
