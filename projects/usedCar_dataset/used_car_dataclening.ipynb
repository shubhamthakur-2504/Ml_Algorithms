## cleaning dataset of used cars

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
%matplotlib inline
```

```python
data=pd.read_csv("cardekho_imputated.csv")
data.head()
```

```python
print(data.columns)
data.head()
```

```python
data=data.drop(columns='Unnamed: 0',axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns='car_name', axis=1)
data.head()
```

```python
len(data['brand'].unique())
```

```python
brand_encoder = LabelEncoder()
model_encoder = LabelEncoder()

data['brand'] = brand_encoder.fit_transform(data['brand'])
data['model'] = model_encoder.fit_transform(data['model'])
data.head()
```

```python
ohe = OneHotEncoder(drop='first')
col_transformer = ColumnTransformer(transformers=[('ohe', ohe , ['seller_type', 'fuel_type', 'transmission_type'])], remainder='passthrough')
data2 = col_transformer.fit_transform(data)
data = pd.DataFrame(data2)
data
```

```python
data.to_csv("cleaned_data.csv", index=False)
```
